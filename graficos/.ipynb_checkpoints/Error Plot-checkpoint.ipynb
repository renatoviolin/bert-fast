```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
data = open('error.txt')
```

```python
lines = []
for l in data:
    lines.append(float(l.replace('\n','').split()[7]))
```

```python
x = np.arange(len(lines))
```

```python
fig = plt.figure(figsize=(16,12))
plt.plot(x, lines)
plt.show()
```
